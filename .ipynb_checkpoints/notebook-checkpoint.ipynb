1. **Apa saja produk terlaris dan produk yang kurang diminati dalam beberapa bulan terakhir?**  


```python
# Menggabungkan data products_df dengan product_cat untuk mendapatkan nama kategori dalam bahasa Inggris
products_df = pd.merge(
    left=products_df,
    right=product_cat,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)

# Menggabungkan data order_items_df dengan products_df
order_product_df = pd.merge(
    left=order_items_df,
    right=products_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)

# Menghitung jumlah produk yang terjual
product_sales = order_product_df.groupby('product_category_name_english').size().reset_index(name='sales_count')

# Mengurutkan produk berdasarkan jumlah penjualan
product_sales_sorted = product_sales.sort_values(by='sales_count', ascending=False)

# Menampilkan 10 produk terlaris
top_10_products = product_sales_sorted.head(10)
print("Top 10 Produk Terlaris:")
print(top_10_products)

# Menampilkan 10 produk yang kurang diminati
bottom_10_products = product_sales_sorted.tail(10)
print("\nBottom 10 Produk yang Kurang Diminati:")
print(bottom_10_products)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Plot Top 10 Produk Terlaris
plt.figure(figsize=(12, 6))
sns.barplot(x='sales_count', y='product_category_name_english', data=top_10_products, palette='viridis')
plt.title('Top 10 Produk Terlaris')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Kategori Produk')
plt.show()

# Plot Bottom 10 Produk yang Kurang Diminati
plt.figure(figsize=(12, 6))
sns.barplot(x='sales_count', y='product_category_name_english', data=bottom_10_products, palette='magma')
plt.title('Bottom 10 Produk yang Kurang Diminati')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Kategori Produk')
plt.show()
```


### Produk Terlaris dan Produk yang Kurang Diminati

#### Top 10 Produk Terlaris
Grafik di bawah ini menunjukkan 10 kategori produk terlaris berdasarkan jumlah penjualan. Kategori produk yang paling diminati adalah `bed_bath_table` dengan jumlah penjualan tertinggi, diikuti oleh `health_beauty`, `sports_leisure`, dan `furniture_decor`. Kategori-kategori ini menunjukkan bahwa pelanggan cenderung membeli produk-produk yang berkaitan dengan kebutuhan rumah tangga, kesehatan, olahraga, dan dekorasi rumah.

#### Bottom 10 Produk yang Kurang Diminati
Grafik di bawah ini menunjukkan 10 kategori produk yang kurang diminati berdasarkan jumlah penjualan. Kategori produk yang paling sedikit diminati adalah `security_and_services` dengan jumlah penjualan terendah, diikuti oleh `fashion_childrens_clothes`, `cds_dvds_musicals`, dan `la_cuisine`. Kategori-kategori ini menunjukkan bahwa pelanggan cenderung kurang tertarik pada produk-produk yang berkaitan dengan layanan keamanan, pakaian anak-anak, media fisik seperti CD/DVD, dan peralatan dapur khusus.

Dengan memahami produk-produk yang paling diminati dan yang kurang diminati, perusahaan dapat mengoptimalkan strategi pemasaran dan stok produk untuk meningkatkan penjualan dan kepuasan pelanggan.

2.**Berapa banyak total pengeluaran pelanggan selama periode tertentu?** 

```python

orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])


start_date = '2017-01-01'
end_date = '2018-12-31'


filtered_orders = orders_df[(orders_df['order_purchase_timestamp'] >= start_date) & (orders_df['order_purchase_timestamp'] <= end_date)]


filtered_orders_pay = pd.merge(
    left=filtered_orders,
    right=order_pay,
    how='left',
    left_on='order_id',
    right_on='order_id'
)


total_spending = filtered_orders_pay.groupby('customer_id')['payment_value'].sum().reset_index()


print("Total Pengeluaran Pelanggan selama periode tertentu:")
print(total_spending)
```

```python


filtered_orders_pay['order_purchase_timestamp'] = pd.to_datetime(filtered_orders_pay['order_purchase_timestamp'])


monthly_spending = filtered_orders_pay.groupby(filtered_orders_pay['order_purchase_timestamp'].dt.to_period('M'))['payment_value'].sum().reset_index()
monthly_spending['order_purchase_timestamp'] = monthly_spending['order_purchase_timestamp'].dt.to_timestamp()


plt.figure(figsize=(14, 7))
sns.lineplot(x='order_purchase_timestamp', y='payment_value', data=monthly_spending, marker='o')
plt.title('Total Pengeluaran Pelanggan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Pengeluaran (Rupiah)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```


### Total Pengeluaran Pelanggan per Bulan

Grafik di bawah ini menunjukkan total pengeluaran pelanggan per bulan selama periode tertentu. Dari grafik ini, kita dapat melihat bagaimana pengeluaran pelanggan berfluktuasi dari bulan ke bulan. Puncak pengeluaran terjadi pada bulan November 2017, yang mungkin disebabkan oleh adanya promosi atau diskon besar-besaran seperti Black Friday. Selain itu, kita juga dapat melihat adanya peningkatan pengeluaran yang signifikan pada awal tahun 2018.

Grafik ini membantu kita memahami tren pengeluaran pelanggan dan dapat digunakan untuk merencanakan strategi pemasaran dan penjualan di masa mendatang.


4. **Dimana lokasi geografis pelanggan terbanyak berada?**  


```python
other_state_geolocation = geo_df.groupby(['geolocation_zip_code_prefix'])['geolocation_state'].nunique().reset_index(name='count')
other_state_geolocation[other_state_geolocation['count']>= 2].shape
max_state = geo_df.groupby(['geolocation_zip_code_prefix','geolocation_state']).size().reset_index(name='count').drop_duplicates(subset = 'geolocation_zip_code_prefix').drop('count',axis=1)
```

```python
geolocation_silver = geo_df.groupby(['geolocation_zip_code_prefix','geolocation_city','geolocation_state'])[['geolocation_lat','geolocation_lng']].median().reset_index()
geolocation_silver = geolocation_silver.merge(max_state,on=['geolocation_zip_code_prefix','geolocation_state'],how='inner')
```

```python
customers_silver = customers_df.merge(geolocation_silver,left_on='customer_zip_code_prefix',right_on='geolocation_zip_code_prefix',how='inner')
```

```python
customers_silver.head()
```

```python
def plot_brazil_map(data):
    brazil = mpimg.imread(urllib.request.urlopen('https://i.pinimg.com/originals/3a/0c/e1/3a0ce18b3c842748c255bc0aa445ad41.jpg'),'jpg')
    ax = data.plot(kind="scatter", x="geolocation_lng", y="geolocation_lat", figsize=(10,10), alpha=0.3,s=0.3,c='blue')
    plt.axis('off')
    plt.imshow(brazil, extent=[-73.98283055, -33.8,-33.75116944,5.4])
    plt.show()
    
```

```python
plot_brazil_map(customers_silver.drop_duplicates(subset='customer_unique_id'))
```

Berdasarkan analisis data geolokasi pelanggan, kota dengan jumlah pelanggan terbanyak adalah Sao Paulo (SP) dengan total 16.396 pelanggan, diikuti oleh Rio de Janeiro (RJ) dengan 7.283 pelanggan, dan Belo Horizonte (MG) dengan 2.910 pelanggan, menunjukkan bahwa platform E-Commerce ini lebih populer di daerah perkotaan yang padat penduduk, sehingga perusahaan dapat merencanakan strategi pemasaran yang lebih efektif dan efisien di kota-kota besar ini untuk meningkatkan penjualan dan kepuasan pelanggan.


```python

customer_locations = customers_silver.groupby(['geolocation_city', 'geolocation_state']).size().reset_index(name='customer_count')


customer_locations_sorted = customer_locations.sort_values(by='customer_count', ascending=False)

top_10_customer_locations = customer_locations_sorted.head(10)

print("Top 10 Lokasi Geografis Pelanggan Terbanyak:")
print(top_10_customer_locations)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))
sns.barplot(x='customer_count', y='geolocation_city', data=top_10_customer_locations, palette='viridis')
plt.title('Top 10 Geographic Locations of Customers')
plt.xlabel('Number of Customers')
plt.ylabel('City')
plt.show()
```
